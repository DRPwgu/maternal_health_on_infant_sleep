## Imputing values on the 
### Very Short Form of the Infant Behavior Questionnaire-Revised (Negative Emotionality dimension) columns

```python
from sklearn.impute import KNNImputer
```

```python
# Function to impute columns for dataframes
def column_knn_imputer(df):
    imputer = KNNImputer()
    columns_to_impute = list(df.columns[df.isnull().any()])
    
    df_imputed = df.copy(deep=True)
    # Impute using fit_transform on the df_imputed
    df_imputed.loc[:, columns_to_impute] = imputer.fit_transform(df.loc[:, columns_to_impute])
    
    return df_imputed
```

```python
df_data = column_knn_imputer(df)
```

```python
# Function to get summary stats
def get_summary_stats_by_columns(df):
    column_name = df.columns
    new_df = pd.DataFrame(index=['Data type', 'Min', '25%', '50%', '75%','Max', 'Mean', 'Median', 'Mode', 'STD', 'Skewness', 'Kurtosis', 'Count'])
    for col in column_name:
        if pd.api.types.is_numeric_dtype(df[col]):
            new_df[col.upper()] = [df[col].dtype, df[col].min(), df[col].quantile(.25), df[col].quantile(.5), df[col].quantile(.75),df[col].max(), df[col].mean(), df[col].median(),
                                   df[col].mode()[0], df[col].std(), df[col].skew(), df[col].kurt() , df[col].count()]
    return new_df
```

```python
# importing library to run the test
from scipy.stats import chi2_contingency
```

```python
# Function to run the test in all categorical variables
def chi2_analisys(df, columns, target):
    """
    Arguments are a dataframe and the name of the Dependent variable as a string
    :param dataframe:
    :param dep_var:
    :return:
    """
    data_to_add = []
    for col in columns:
        temp_crosstab = pd.crosstab(df[target], df[col])
        temp_result = chi2_contingency(temp_crosstab)
     
        if temp_result[1] < 0.05:
            # print(f'Dependent Var: {dep_var} ---- Independent Var: {col} ----- p-value={round(temp_result[1], 10)} -> We will reject the NULL hypothesis.')
             data_to_add.append([target, col, round(temp_result[1], 10), 'Reject'])
        else:
             # print(f'Dependent Var: {dep_var} ---- Independent Var: {col} ----- p-value={round(temp_result[1], 10)} -> We will not reject the NULL hypothesis.')
            data_to_add.append([target, col, round(temp_result[1], 10), 'Accept'])
    new_df = pd.DataFrame(data_to_add, columns=['Dependent', 'Independent', 'p-value', 'NULL Hypothesis'])
    return new_df.sort_values(by='NULL Hypothesis').reset_index(drop=True)
```

```python
# NULL hypothesis -> the variables have no association or correlation amongst them.
chi2_analisys(df_data, q_col, 'Sex')
```
